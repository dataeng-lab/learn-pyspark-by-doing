```python
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder.appName("PySparkExample").getOrCreate()

# Step 1: Read CSV (transformation, no execution yet)
df = spark.read.csv("/content/sample_data/california_housing_test.csv", header=True)

# Step 2: Apply filter (still a transformation, Spark just records it)
filtered = df.filter(df['latitude'] > 1)

# Step 3: Action â€“ triggers execution
filtered.show()
```
